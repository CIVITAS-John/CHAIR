```python
import json
from datetime import datetime


# Define a function to process each .txt file and format it into the desired JSON format
def process_text_file(file_id, file_content):
    # Trim leading/trailing spaces and split into lines
    lines = file_content.strip().split("\n")

    # Generate the chunk ID from the file name (excluding the extension)
    chunk_id = file_id.split(".")[0]

    # Prepare the container for this chunk
    chunk_data = {
        "ID": chunk_id,
        "Start": "1970-01-01 00:00:00",
        "End": "1970-01-01 00:00:00",
        "AllItems": [],
    }

    # Variables to keep track of speaker and content
    current_speaker = None
    item_id = 1

    # Process lines to extract speaker and their content
    for line in lines:
        line = line.strip()  # Remove any extra whitespace
        if not line:  # Skip empty lines
            continue

        if line.endswith(":"):  # Identifying the speaker line
            current_speaker = line.replace(":", "")
        else:
            if current_speaker:
                # Add this speaker's content as a new item
                item = {
                    "ID": f"{chunk_id}-{item_id}",
                    "UserID": current_speaker,
                    "Time": "1970-01-01 00:00:00",
                    "Content": line,
                }
                chunk_data["AllItems"].append(item)
                item_id += 1

    return chunk_data
```

```python
import os


def process_text_files_in_folder(root_folder):
    """Process all .txt files in the given folder and transform them into the desired JSON format."""

    def load_local_file(file_path):
        """Load the content of a local file."""
        with open(file_path, "r", encoding="utf-8") as file:
            return file.read()

    # Search for all .txt files in the specified folder
    txt_files = [f for f in os.listdir(root_folder) if f.endswith(".txt")]

    # Dictionary to store the final JSON data
    final_data = {}

    # Process each .txt file found
    for txt_file in txt_files:
        file_path = os.path.join(root_folder, txt_file)
        file_content = load_local_file(file_path)
        processed_file = process_text_file(
            txt_file, file_content
        )  # Reuse the existing function
        final_data[processed_file["ID"]] = processed_file

    # Save the resulting JSON to a file
    output_file_path = os.path.join(root_folder, "data-example.json")
    with open(output_file_path, "w", encoding="utf-8") as f:
        json.dump(final_data, f, indent=4)

    return output_file_path


# Example usage:
root_folder_path = "./"
output_json_path = process_text_files_in_folder(root_folder_path)
output_json_path
```
